EJ 6: Explique cuál es diferencia entre localización de objetos y clasificación de imágenes. Muestre ejemplos de ello.

Aunque puedan parecer similares la localizacion de objetos y la clasificacion de imagenes se diferencian en 3 puntos fundamentales, su salida o resultado, la complejidad y las aplicaciones.

La **localizacion de objetos** identifica la precencia y ubicacion de un objeto dentro de una imagen, y le asigna un bounding box que lo delimita. Se puede usar para la deteccion de personas en camaras de vigilancia, autos en la calle u objetos en una sala.
El ejercicio anterior es un caso de localizacion de objetos.

Por su parte la **clasificacion de imagenes**, clasifica una imagen dentro de una serie de categorias predefinidas, asignandole una etiqueta.
Los ejemplos mas comunes pueden ser perro o gato, digitos escritos a mano, piedra papal o tijera,etc.

Retomando lo señalado en cuanto a las diferencias:
- *Salida*: La localizacion de objetos provee un bounding box delimitando el objeto detectado, la clasificacion de imagenes asigna una etiqueta de acuerdo a su prediccion.
- *Complejidad*: por lo general la deteccion de imagenes requiere delimitar donde se ubica el objeto por lo que es mas compleja que una clasificacion sobre toda una imagen.
- *Aplicaciones*: para la deteccion de objetos se destacan segmentacion de imagen, realidad aumentada, seguimiento de objetos.
Para la clasificacion de imagenes podemos mencionar su uso en diagnosticos medicos, reconocimiento de productos, filtrada en redes sociales.
